## shutil.move() 기능
---
> shutil.move()는 동일하거나 다른 파일 시스템 내에서 파일을 이동한다. 파일을 이동하기 위해 shutil은 move() 기능, 그것은 호출 os.rename() 목적지가 소스와 동일한 디스크에 있을 때 기능 그렇지 않으면 원본 파일을 대상에 복사한 다음에 삭제한다.
---
```python

import shutil
import os
 
src = '/path/to/src/dir'
dest = '/path/to/dest/dir'
 
files = os.listdir(src)
 
for f in files:
    shutil.move(src + f, dest)
```
> 이 코드는 shutil 모듈을 사용하여 소스 디렉토리(src)의 파일들을 대상 디렉토리(dest)로 이동시키는 것이다
1. os.listdir() 함수를 사용하여 소스 디렉토리에 있는 모든 파일의 이름을 가져온다.
2. shutil.move(src + f, dest)는 src 경로와 f 파일 이름을 결합하여 파일의 전체 경로를 만들고, 해당파일을 dest 경로로 이동시킨다.
